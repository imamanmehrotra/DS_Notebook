```python
df1['Embarked'].value_counts()
```

```python
#Pandas

pd.get_dummies(df1['Embarked'], drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
l=pd.DataFrame(le.fit_transform(df1['Embarked']), columns=['Embarked'])
l
```

```python
#le.fit_transform(df1['Embarked']) -->for reading the transformed data
```

```python
df1.drop('Embarked',axis=1,inplace=True)   #Columns in pandas are rep by 1 and row by 0
```

```python
df1=pd.concat((df1,l),axis=1)
```
